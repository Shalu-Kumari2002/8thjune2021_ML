```python
def preprocess(t):
    words = []
    for sent in nltk.sent_tokenize(t.lower()):    
        ws = nltk.word_tokenize(sent)
        ws = [w for w in ws if w.isalpha() and w not in swe]
       # ws = [w for w in ws if w not in punc]    
        words+=(ws)
    words = ' '.join(words)
    return words
```

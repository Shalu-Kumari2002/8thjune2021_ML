```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('features_30_sec.csv')
```

```python
df1.head()
```

```python
df1.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df1.drop(['label','filename'],axis =1 )
y = df1['label']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.30,random_state = 101)
```

```python
df1['label']
```

```python
model_sv = SVC().fit(xtrain,ytrain)
model_sv.score(xtest,ytest)*100
```

```python
model_Legit = LogisticRegression().fit(xtrain,ytrain)
model_Legit.score(xtest,ytest)*100
```

```python
model_kn = KNeighborsClassifier().fit(xtrain,ytrain)
model_kn.score(xtest,ytest)*100
```

```python
model_randome = RandomForestClassifier().fit(xtest,ytest)
model_randome.score(xtest,ytest)*100
```

```python
#so we can see that the best way to make model here is Random Forest Classifier
```

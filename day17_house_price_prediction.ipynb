# Feature seletion

```python
df.columns 
```

```python
#Feature Matrix
x = df[['Avg. Area Income','Avg. Area House Age','Avg. Area Number of Rooms','Area Population']]  #which are the features of our price making thing 
#label matrix 
y = df['Price']  #this is what we are gonna find 
```

# Data Splitting 

```python
from sklearn.model_selection import train_test_split  #This is that particular library where all this things have been kept
```

```python
#train_test_split actually divides the data into training data , and test data, the size of test data is 25% of the total data_as shown by the parameter test_size
#pehle x to todega to wese hi name do 
#random function shows randomness  like agar nhi diya to bar bar run karne par wo alag alag dikha raha hai
```

```python
train,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state=101)
```

# Linear Regression Modelling

```python
from sklearn.linear_model import LinearRegression
```

```python
#fit wo training data leta hai , or sara distance and line banata hai 
```

```python
#Model Creating AND Training 
#fit is used to seek training data
#this Linear Regresssion applys all its appraoches that is finding eucledian distance and then coefficienst 
#after that let's see
model = LinearRegression().fit(xtrain,ytrain)
```

```python
model 
```

```python
model.coef_
```

```python
coef = pd.DataFrame(model.coef_,columns =['Coeff'],index =x.columns )  #here we are giving names as x.columns as this will provide us with coeff which shows the unit change or increament 
#suppose if i have 1 dollars as income then the house rate will be these much , if 2 dollars then 21.64*2 
#if i want this much house age then price will increase this way
#if i want this much area of rooms 
#if the population of tha area is this much
#then it will calculate based on this coeficients the price 
#so get it na
```

```python
coef
```

```python
#suppose ki income jo hai 1$ to us ghar ka price hona chaiye 21.
#similarly agar ghar ki age ek sal purani hai to 164948 badh jayegi 
#for unit change in features what is the change in label is calculated 
```

# Model Evaluatoion

```python
model.score(xtest,ytest)  #ye hume ye test data kya deta hai dikhayega , matlab actually ye hume ye batayehga ki test data or prediction kitna same hai 
#what it actually does is that it takes the real data that is test data and based on the model it created using the train data predicts the accuracy it reaches
```

```python
round(100*model.score(xtest,ytest),2)  #itni accuracy rahi model ki #rounded of by 2 digits
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error 
```

```python
ypred = model.predict(xtest)  #ye predict hi sabse jyada kam ati hai #it is predicting on the basis of x test data , features the prediction
#it is prediction that if test data has such values then what price my house will take
```

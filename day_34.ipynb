```python
from nltk.chat.util import Chat,reflections
```

```python
reflections  #i m ko you are bana dena 
```

pairs = ((ques,(resp1,resp2)),(ques,(resp1,resp2)),(ques,(resp1,resp2)))   
#pair kya hai tuple jisme question bhi hai or uske multiple responses 

```python
pairs  = (
    (r'i like (.*)',
     (
         'Surely , you can like %1',
         'why, do you like %1'
     )
    ),
    (r'will you please (.*)',
     (
         'yeah, sure',
         'no, i can not do this'
     )
    ),
     (r'Hello',
     (
         'HI madam',
         'can u just get lost '
     )
    ),
     (r'(.*)python(.*)',
     (
         'python is an interpreted language ',
         'u have a good questio'
     )
    ),
)
```

```python
cb = Chat(pairs,reflections) #questions ke pattern banayenge or response banayenge 
```

```python
cb # small chat bot service 
```

```python
import pyttsx3  #to reply in audio
```

```python
speng = pyttsx3.init()  #speech engine initialisation
```

```python
speng.setProperty('rate',120)
```

```python
speng.say('hello I am speaking')
speng.runAndWait()
```

```python
import speech_recognition as sr
```

```python
r = sr.Recognizer()
```

```python
while True:
    with sr.Microphone() as source:  # here i set my microphone as my recording device
        r.adjust_for_ambient_noise(source)  #clearing the noise in the source
        audio = r.listen(source)            ##record what i said
        try:
            que = r.recognize_google(audio)  #jo bhi tuumne suna na babe wo google se pucho ki text me kya meaning banta hai 
            if que == 'exit':
                break
            print('you said:',que)
            resp = cb.respond(que) 
            print('Bot said:',resp)
            speng.say(resp)
            speng.runAndWait()
            print('-'*20)#or jo banta hai ise question banao or answer karo 
        except Exception as e:               # or agar koi error hai to rehne de 
            print('error')
            print(e)
```

```python
#importing the libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading the data 
df = pd.read_csv('50_Startups.csv')
```

```python
#Creating dummy data 
x = 1.5*np.random.normal(0,1,30)               #creating normal data 
x
x.shape
x
```

```python
plt.plot(x)
```

```python
plt.semilogx(x)
```

```python
#i want y which is in the form of -- b0x^0+b1x^1+b2x^2...
```

```python
y = x - 3*(x**2) + 0.6*(x**3) + np.random.normal(0,1,30)
y
```

```python
plt.scatter(x,y,s=25,c = 'r')  #its looks like curved data
#now if we want to increase the spread 
#i m changing y for that -- np.random.normal(-2,2,30) - np.random.normal(-4,4,30)
#the more the values range in normal , the more th spread
```

```python
x = x[:,np.newaxis]
x
```

```python
y = y[:,np.newaxis]
y
```

```python
x.shape
```

```python
#importing Linear regression 
from sklearn.linear_model import LinearRegression 
```

```python
#now i want to make a model and pass x,y to fit in linear regression model 
model1 = LinearRegression().fit(x,y)

#we need to pass 2D array using np.newaxis
```

```python
#since wea are working on training data , we are not seperating the data into training and test, so now we will predict y , by fitting x on it
#using the same data in trainng and test or evaluation is called training performance
ypred = model1.predict(x)
```

```python
model1.score(x,y)*100
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y,c='b')
plt.plot(x,ypred,c='g')
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x,y,s= 25,c = 'r')
plt.plot(x,ypred,c='g')
plt.grid
```

```python
#In the above fig we can see that this line is not at all fitting our at our data , so this is not working at all
```

```python
#now i want to preprocess my data 
from sklearn import preprocessing as pp
```

```python
x
```

```python
#now i will make my normal data into polynomial data 
x_poly = pp.PolynomialFeatures(1).fit_transform(x)
x_poly
x_poly.shape
```

```python
y.shape
y
```

```python
#now i wil make model based on this preprocessed data
x_poly = pp.PolynomialFeatures(1).fit_transform(x)
x_poly
model_d1 = LinearRegression().fit(x_poly,y)
ypred_d1 = model_d1.predict(x_poly)
print(model_d1.score(x_poly,y)*100)
plt.figure(figsize=(10,6))
plt.scatter(x,y,s=25,c='g')
plt.plot(x,ypred_d1,c='r')
plt.grid()
```

```python
model_d1.score(x_poly,y)*100
```

```python
x_poly2 = pp.PolynomialFeatures(2).fit_transform(x)
model_d2 = LinearRegression().fit(x_poly2,y)
ypred_d2 = model_d2.predict(x_poly2)
print(model_d2.score(x_poly2,y)*100)
plt.figure(figsize=(10,6))
plt.scatter(x,y,s=25,c='g')
plt.scatter(x,ypred_d2,c='r')
plt.grid()
```

```python
x_poly3 = pp.PolynomialFeatures(3).fit_transform(x)
model_d3 = LinearRegression().fit(x_poly3,y)
ypred_d3 = model_d3.predict(x_poly3)
print(model_d3.score(x_poly3,y)*100)
plt.figure(figsize=(10,6))
plt.scatter(x,y,s=25,c='g')
plt.scatter(x,ypred_d3,c='r')
plt.grid()
```

```python
x_poly4 = pp.PolynomialFeatures(4).fit_transform(x)
model_d4 = LinearRegression().fit(x_poly4,y)
ypred_d4 = model_d4.predict(x_poly4)
print(model_d4.score(x_poly4,y)*100)
plt.figure(figsize=(10,6))
plt.scatter(x,y,s=25,c='g')
plt.scatter(x,ypred_d4,c='r')
plt.grid()
```

```python
x_poly5 = pp.PolynomialFeatures(5).fit_transform(x)
model_d5 = LinearRegression().fit(x_poly5,y)
ypred_d5 = model_d5.predict(x_poly5)
print(model_d5.score(x_poly5,y)*100)
plt.figure(figsize=(10,6))
plt.scatter(x,y,s=25,c='g')
plt.scatter(x,ypred_d5,c='r')
plt.grid()
```

```python
#algo method 
score = []
for i in range(1,10):
    x_poly_i = pp.PolynomialFeatures(i).fit_transform(x)
    model_d_i = LinearRegression().fit(x_poly_i,y)
    ypred_d_i = model_d_i.predict(x_poly_i)
    score.append(model_d_i.score(x_poly_i,y)*100)

```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,10),score,marker= 'o')
plt.grid()
```

##### Treatment of missing data 

```python
## data ko remove kardo 
#jese row ko 
#ya column ko remove kardo 
```

```python
#agar 50 or jyda data hi nhi hai to pura remove kardo 
#agar 2-8 percent missing hai to usme hum khud bhar sakte hai 

#rows ko remove kardene se wo banjayega data loss #to wahan kuch or bhar dete hai like kuch logic ke throgh ab wo logic depend karta hai 
#ki baki data kese hai 
#mathematical operation karna hai to usme to chaiye hoga ki kya karna hai 
#to wo row ko remove kardo 
#agar 10% data hai sirf to pura remove kardo 
```

```python
df.dropna()   #remove missing values #there is three default values #axis = 0(for index),1(columns ),thresh = isse kam ya jyada ya iss condition se match karta hai to remove kar , inplace = True mane asli data me  change kar 
```

```python
df.dropna(axis = 1)  #galat #removing rows 
```

```python
df.dropna(axis 0,thresh =2)  #thresh kya karta hai ye remove nhi karta 
```

```python
df.dropna(axis = 0)
```

```python
#hum mean values se fill up kar dete hai 
#thresh kya karta hai ki mano 4 column hai or usme mujhe wo row hi rakhna hai jisme jisme 3 column me data ho kamsekam to ,to me thresh me 3 
df.dropna(axis = 0, thresh = 3)
```

```python
d1 = pd.DataFrame({'name':['Rahul','vishesh','Kanika'],'age':[23,22,20],'uid':[101,102,103]})
```

```python
d1
```

```python
d2 = pd.DataFrame({'subject':['python','python','cpp','php'],'uid':[103,101,104,105]})
```

```python
d2
```

```python
#now suppose i want ki iss d1 or d2 me se hume ek common info chaiye jo ki hai uid or phir isse se mujhe merge karna hai 
pd.merge(d1,d2, on = 'uid')
```

```python
pd.merge(d1,d2,)
```

```python
pd.merge(d1,d2, on = 'uid', how='outer')  #ab use karenge how = outer matlab ki hum sara le ayenge ir jo nhi hai use bhi dal denge 
```

```python
#How = left matlab ki left wale ka data to hona hi chaiye 
# how = right matlab right wale ka data to ana hi chaiye 
pd.merge(d1,d2, on = 'uid', how='left')
```

```python
pd.merge(d1,d2, on = 'uid', how='right')
```

```python

app = tk.Tk()
app.geometry("350x350")
app.configure(bg = 'pink')
app.title('The Price predictor')

m = pd.read_pickle('house_price_predictor.pkl')

income_amnt = tk.Variable(app)
age_of_house = tk.Variable(app)
no_of_rooms = tk.Variable(app)
input_population_of_that_area = tk.Variable(app)
m = pd.read_pickle('house_price_predictor.pkl')

price_of_the_house =tk.Variable(app)
price_of_the_house.set('')







    

def price_output():
    inc = eval(income_amnt.get())
    age =eval(age_of_house.get())
    room = eval(no_of_rooms.get())
    pop = eval(input_population_of_that_area.get())
    query = pd.DataFrame({'income':[inc],'house_age':[age],'rooms':[room],'population':[pop]})
    round(m.predict(query)[0],2)
    price_of_the_house.set(round(m.predict(query)[0],2))


entry1 = Entry(app,textvariable = income_amnt,bg='white',fg='black',bd =3,font=('calibre',10,'normal'),width = 18)  #textvariable = input_income font=('calibre',10,'normal')
entry1.place(x=200,y=50)
lb1 = tk.Label(app,text = 'input_income',font =('calibre',10,'bold')).place(x = 5, y = 50)
entry2 = Entry(app,textvariable = age_of_house,bg='white',fg='black',bd =3,font=('calibre',10,'normal'),width = 18)#textvariable = input_age_of_house
entry2.place(x=200,y=100,)
lb2 = tk.Label(app,text = 'input_age_of_house',font =('calibre',10,'bold')).place(x = 5, y = 100)
entry3 = Entry(app,textvariable = no_of_rooms ,bg='white',fg='black',bd =3,font=('calibre',10,'normal'),width = 18) #textvariable = input_number_of_rooms
entry3.place(x=200,y=150)
lb3 = tk.Label(app,text = 'input_no_of_rooms',font =('calibre',10,'bold')).place(x = 5, y = 150)
entry4 = Entry(app,textvariable =input_population_of_that_area,bg='white',fg='black',bd =3,font=('calibre',10,'normal'),width = 18)
entry4.place(x=200,y=200)
lb4 = tk.Label(app,text = 'input_population',font =('calibre',10,'bold')).place(x = 5, y = 200)
btn1 = tk.Button(app,command =lambda:price_output(), text = 'Calculate the house price :').place(x=5,y=250)
lb5 = tk.Label(app,textvariable =price_of_the_house,text = 'Price of the house ',font =('calibre',10,'bold'),).place(x = 200, y = 250)





app.mainloop()
```

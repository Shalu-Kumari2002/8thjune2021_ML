```python
inert = []
for i in range(2,11):
    model_i = KMeans(n_clusters=i).fit(dummy)
    inert.append(model_i.inertia_)
```

# SOM for Image Compression

```python
img = plt.imread('rose.jpg')
```

```python
plt.imshow(img)
```

```python
img.shape  #image is a 3d matrix -- and three colors - r y b 
```

```python
img.T
```

```python
img.dtype
```

Before csize conversion let's chnage the shape

```python
import numpy as np
```

```python
pix = img.reshape((img.shape[0]*img.shape[1],3)) # 3 is number of planes 
```

```python
pix = pix.astype(float)/255
```

```python
pix.shape
```

```python
from minisom import MiniSom
```

```python
som = MiniSom(x=3,y=3,input_len = 3,sigma = 0.01, learning_rate=0.1)
```

```python
som.random_weights_init(pix)
```

```python
i_w = som.get_weights().copy()
```

```python
i_w
```

```python
som.train_random(pix, 100,verbose=True)
```

```python
data_qt = som.quantization(pix)  #this is the utput of compressed data 
```

```python
data_qt
```

```python
rec_img = data_qt.reshape(img.shape)
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(rec_img)
plt.show()
```

```python
plt.imsave('rec.jpg',rec_img)
```

```python
rec_img.shape
```
